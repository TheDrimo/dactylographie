```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = open("./dactylo", "r")
l = f.read()
for c in ["[","]","{","}"]:
    l = l.replace(c, "")
l = l.split(",\"")
precision, cpm, mpm, fluidite = [], [], [], []
for i in l:
    if "precision" in i:
        i = i.replace("\"", "")
        i = i.replace("precision:", "")
        precision.append(float(i))
    if "cpm" in i:
        i = i.replace("\"", "")
        i = i.replace("cpm:", "")
        cpm.append(float(i))
    if "mpm" in i:
        i = i.replace("\"", "")
        i = i.replace("mpm:", "")
        mpm.append(float(i))
    if "fluidite" in i:
        i = i.replace("\"", "")
        i = i.replace("fluidite:", "")
        fluidite.append(float(i))
data = {"precision":precision, "cpm":cpm, "mpm":mpm, "fluidite":fluidite}
df = pd.DataFrame(data)
```

```python
fig, axes = plt.subplots(3, sharex=True, figsize=(15, 10))
for a in range(0, len(axes)) : 
    axes[a].plot(df.index, df[df.columns[a]])
    axes[a].set_ylabel(df.columns[a])
    axes[a].grid()
fig.suptitle('Ma progression, essai aprs essai, en dactylographie')
plt.show()
```
